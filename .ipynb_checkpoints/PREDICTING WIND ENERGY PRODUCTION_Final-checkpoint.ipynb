### 1.7. Feature transformation

Althought we have many variables, it is understandable that the variables u100 (100 metre U wind component) and v100 (100 metre V wind component) give us the most valuable information. However, the modullus is more valuable, as measures the wind speed at 100 meters high. Therefore, we would add this as a new variable to our data. 

```python
v_wind_columns = wind_ava.columns[wind_ava.columns.str.startswith('v100.')]
u_wind_columns = wind_ava.columns[wind_ava.columns.str.startswith('u100.')]

# Iterate through pairs of u and v columns
for u_col, v_col in zip(u_wind_columns, v_wind_columns):
    # Extract u and v components
    u_component = wind_ava[u_col]
    v_component = wind_ava[v_col]

    # Calculate wind speed
    wind_speed_col = f"wind_speed{u_col.split('.')[1]}"
    wind_ava[wind_speed_col] = np.sqrt(u_component**2 + v_component**2)

# Display the modified DataFrame with new wind speed columns
wind_ava.head()
```

As can be observed, this new variable againts the response variable follows a linear relation, so it will give us valuable information when modelling the predictor.

### 2.2. Metric

The choice of which metric to use for evaluating a regression model depends on the specific characteristics of the problem and our goals. As a summary of the benefits and drawbacks of the most commonly used regression metrics:

- Mean Squared Error (MSE) or Root Mean Squared Error (RMSE): These are often preferred when large errors should be penalized more heavily, and the distribution of errors is approximately normal. However, if your data contains outliers, MSE and RMSE can be significantly affected.

- Mean Absolute Error (MAE): These are more robust to outliers. If your dataset has a skewed distribution or contains outliers, MAE might be more appropriate.

Therefore, we will evaluate the outliers of our objetive variable to understand which metric is better on this problem. As observed, the data is rigth skewed, with many outliers on the rigth section, therefore, a MAE metric would affect less to these errors when predicting outliers.

## 3. Preprocessing 

### 3.1. Feature selection

In predictive modeling, not all columns or features may be necessary, and including irrelevant or redundant ones poses several challenges. The curse of dimensionality arises when a large number of features contribute to overfitting and increased computational complexity, requiring more data for reliable estimates. Superfluous features hinder model interpretability and introduce noise, potentially leading to issues with data quality, missing values, and outliers. Feature selection becomes crucial in mitigating these challenges, helping to identify and retain the most informative features for building simpler and effective predictive models.

In the initial stage of preprocessing, we prioritized feature selection as the first step, anticipating the subsequent use of the iterative imputer and aiming to optimize computational efficiency. To kickstart this process, we employed SimpleImputer for initial imputation, employing the default mean as the descriptive statistic to handle missing values. Concurrently, we introduced StandardScaler to prepare the data for feature selection. Our feature selection strategy involved leveraging SelectKBest, a sklearn method implementing univariate feature selection, which selects optimal features based on statistical tests. Within this phase, our objective was to determine the ideal parameters for SelectKBest, and encompasse the feature selection score function and the optimal number of features to select. Additionally, SelectKBest provided a valuable output – a mask or integer index – indicating the selected features. 

```python
#We set a seed for reproducibility, althougth the split is predefined, some steps in all the pipelines defined may have some randomness
random_state= 100516919

# DATA
X = X_train
y = y_train

# Define the holdout split: train-train and train-validation 
indices= ([-1] * (len(X_train_train))) + ([0] * (len(X_train_validation)))
inner= PredefinedSplit(indices)


pipeline_feature_selection = Pipeline([
    ('imputer', SimpleImputer()), 
    ('scaler', StandardScaler()),
    ('feature_selection', SelectKBest()),  
    ('knn', KNeighborsRegressor(n_jobs=-1))
])

# Define the hyperparameter grid for GridSearchCV
param_grid_fs = {
    'feature_selection__score_func': [f_regression, mutual_info_regression],
    'feature_selection__k': [100,150,250,300,350] 
}


# Use GridSearchCV to find the best combination of parameters
grid_search = GridSearchCV(pipeline_feature_selection, param_grid_fs, cv=inner, scoring='neg_mean_absolute_error', n_jobs=-1)
grid_search.fit(X, y)

# Get the best model from the grid search
best_model = grid_search.best_estimator_
```

```python
best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")
```

```python
# Access the negative mean absolute error 
best_score = grid_search.best_score_
mae = abs(best_score)
print("Mean Absolute Error for the Best Model:", mae)
```

```python
# Access the selected features
selected_features = best_model.named_steps['feature_selection'].get_support()

# Print the indices of selected features
selected_feature_indices = np.where(selected_features)[0]
print(f"Indices of selected features: {selected_feature_indices}")
```

### 3.2. Imputation


Imputation is a crucial step in the data preprocessing pipeline, particularly when dealing with missing values in a dataset. The presence of missing values can hinder the performance of machine learning algorithms, and imputation methods aim to address this challenge by providing estimated values for the missing entries. In sklearn, various imputation strategies can be employed, each with its own set of advantages and drawbacks. The SimpleImputer offers straightforward strategies such as mean, median, or most frequent value imputation. It is computationally efficient and easy to implement but may oversimplify the data structure. The IterativeImputer is a more sophisticated approach that iteratively estimates missing values based on the relationships observed in the non-missing entries. This method can handle complex dependencies but is computationally more intensive. The KNNImputer utilizes the k-nearest neighbors algorithm to impute missing values, leveraging the information from nearby data points. This method is effective when there are patterns in the data, but it may be sensitive to the choice of k. 
The Mean Absolute Error will be computed for the three models with different imputations, and the one with the lowest value will be selected.

```python
random_state= 100516919

# Load a sample dataset for demonstration
X = X_train_selected
y = y_train

# Define the holdout split
indices = ([-1] * len(X_train_train)) + ([0] * len(X_train_validation))
inner = PredefinedSplit(indices)

# Create a new pipeline with imputation after feature selection
pipeline_simple_imputer = Pipeline([
    ('imputer', SimpleImputer()), 
    ('scaler', StandardScaler()),  
    ('knn', KNeighborsRegressor(n_jobs=-1))
])

# Define the hyperparameter grid for GridSearchCV
param_grid_si = {
    'imputer__strategy': ['mean', 'median', 'most_frequent']
}

# Use GridSearchCV to find the best parameter
grid_search = GridSearchCV(pipeline_simple_imputer, param_grid_si, cv=inner, scoring='neg_mean_absolute_error', n_jobs=-1)
grid_search.fit(X, y)

# Get the best model from the grid search
best_model = grid_search.best_estimator_

# Access the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Access the negative mean absolute error (score) for the selected imputation strategy
best_score = grid_search.best_score_
mae = abs(best_score)
print("Mean Absolute Error for the Best Model:", mae)
```

```python
# train-train data
X = X_train_train_selected
y = y_train_train

# Create a new pipeline with imputation after feature selection
pipeline_iterative_imputer = Pipeline([
    ('imputer', IterativeImputer()), 
    ('scaler', StandardScaler()),  
    ('knn', KNeighborsRegressor(n_jobs=-1))
])

pipeline_iterative_imputer.fit(X, y)

# Make predictions on the validation set
y_pred = pipeline_iterative_imputer.predict(X_train_validation_selected)

# Calculate and print the Mean Absolute Error (MAE)
mae = mean_absolute_error(y_train_validation, y_pred)
print(f"Iterative imputer MAE: {mae}")
```

### 3.3. Scaling

Once we have completed feature selection and imputation in our preprocessing pipeline, the next essential step is scaling. Scaling is particularly crucial for distance-based models, like k-nearest neighbors, as it ensures that all features contribute uniformly to the distance calculations. Without scaling, features with larger scales might dominate the distance metric, leading to biased results. Interestingly, tree-based models, such as decision trees and random forests, are inherently insensitive to feature scales due to their structure, as they make decisions based on hierarchical splits and do not rely on distance metrics. However, for the sake of consistency and to simplify the overall preprocessing workflow, we may still choose to perform scaling on the entire dataset.
Now, considering three common scaling methods in sklearn, each comes with distinct trade-offs. The StandardScaler robustly standardizes features by removing the mean and scaling to unit variance, but it can be sensitive to outliers. The MinMaxScaler preserves relative relationships by scaling features to a specified range, offering robustness against outliers but potentially struggling with skewed data. The RobustScaler is outlier-resistant, scaling based on interquartile range, but may falter with heavily skewed distributions. 
To test which is better for our dataset, the MAE of each model will be calculated, and the one with the minimum value will be selected.

```python
X = X_FS_imputed_train_train
y = y_train_train

# Create a new pipeline with scaling after feature selection and imputation
pipeline_standard_scaler = Pipeline([
    ('scaler', StandardScaler()), 
    ('knn', KNeighborsRegressor(n_jobs=-1))
])

pipeline_standard_scaler.fit(X, y)

# Make predictions on the validation set
y_pred = pipeline_standard_scaler.predict(X_FS_imputed_train_val)

# Calculate and print the Mean Absolute Error (MAE)
mae = mean_absolute_error(y_train_validation, y_pred)
print(f"Standard scaler MAE: {mae}")
```

```python
X = X_FS_imputed_train_train
y = y_train_train

# Create a new pipeline with scaling after feature selection and imputation
pipeline_minmax_scaler = Pipeline([
    ('scaler', MinMaxScaler()), 
    ('knn', KNeighborsRegressor(n_jobs=-1))
])

pipeline_minmax_scaler.fit(X, y)

# Make predictions on the validation set
y_pred = pipeline_minmax_scaler.predict(X_FS_imputed_train_val)

# Calculate and print the Mean Absolute Error (MAE)
mae = mean_absolute_error(y_train_validation, y_pred)
print(f"MinMax scaler MAE: {mae}")
```

```python
# Create a new pipeline with scaling after feature selection and imputation
pipeline_robust_scaler = Pipeline([
    ('scaler', RobustScaler()), 
    ('knn', KNeighborsRegressor(n_jobs=-1))
])

pipeline_robust_scaler.fit(X, y)

# Make predictions on the validation set
y_pred = pipeline_robust_scaler.predict(X_FS_imputed_train_val)

# Calculate and print the Mean Absolute Error (MAE)
mae = mean_absolute_error(y_train_validation, y_pred)
print(f"Robust scaler MAE: {mae}")
```

### 4.6. Optuna

The objective function, defined within the objective function, dynamically creates and evaluates models based on the different methods previously done: decision trees, k-nearest neighbors, random forests, and gradient boosting. Optuna's Tree-structured Parzen Estimator (TPE) sampler is utilized to guide the search efficiently. The objective is to maximize the negative mean absolute error (MAE) obtained through cross-validation on the training dataset.

The benefits of using Optuna for hyperparameter tuning are significant. By automating the search process, Optuna efficiently explores the hyperparameter space, determining optimal configurations that maximize the model's performance. This approach is far more effective and less prone to human biases compared to manual tuning. Optuna's intelligent sampling algorithms, such as TPE, help in selecting promising hyperparameter values, speeding up the optimization process. 

```python
import optuna
def objective(trial):
    method = trial.suggest_categorical('method', ['tree', 'knn', 'random_forest', 'gradient_boosting'])
    if method== 'tree':
        max_depth=trial.suggest_int('max_depth', 2, 20)
        min_samples_split= trial.suggest_int('min_samples_split', 50, 150, step=10)
        params = {'max_depth': max_depth, 'min_samples_split': min_samples_split}
        regr = DecisionTreeRegressor(random_state=random_state, **params)
        
    elif method=='knn':
        n_neighbors= trial.suggest_int('n_neighbors',1,50)
        weights = trial.suggest_categorical('weights', ['uniform', 'distance'])
        algorithm = trial.suggest_categorical('algorithm', ['auto', 'ball_tree', 'kd_tree', 'brute'])
        metric= trial.suggest_categorical('metric', ['minkowski','euclidean', 'chebyshev', 'manhattan'])

        if metric == 'minkowski':
            p= trial.suggest_float('p', 1.0,3.0)
            params = {'n_neighbors': n_neighbors,'weights': weights,'algorithm': algorithm, 'metric': metric, 'p': p }
        else:
            params = {'n_neighbors': n_neighbors,'weights': weights,'algorithm': algorithm, 'metric': metric}
        
        regr = KNeighborsRegressor(**params)

    elif method=='random_forest':
        max_depth=trial.suggest_int('max_depth', 2, 20)
        min_samples_split= trial.suggest_int('min_samples_split', 50, 150,step= 10)
        n_estimators= trial.suggest_int('n_estimators', 50, 150,step=10)
        max_features= trial.suggest_float('max_features', 0.2, 1.0)
        params = {'max_depth': max_depth, 'min_samples_split': min_samples_split,'n_estimators': n_estimators, 'max_features': max_features }
        regr = RandomForestRegressor(random_state=random_state, **params)

    elif method=='gradient_boosting':
        max_depth=trial.suggest_int('max_depth', 2, 20)
        min_samples_split= trial.suggest_int('min_samples_split', 50, 150, step=10)
        n_estimators= trial.suggest_int('n_estimators', 50, 150, step=10)
        learning_rate= trial.suggest_float('learning_rate', 0.01, 0.21, step=0.02)
        params = {'max_depth': max_depth, 'min_samples_split': min_samples_split,'n_estimators': n_estimators, 'learning_rate': learning_rate }
        regr = GradientBoostingRegressor(random_state=random_state, **params)

    scores= cross_val_score(regr, X_train, y_train, scoring= 'neg_mean_absolute_error',
                            n_jobs=-1, cv=inner)
    return scores.mean()

budget=15
sampler= optuna.samplers.TPESampler(seed=random_state)
study = optuna.create_study(direction='maximize', sampler=sampler)
start_time = time.time()

study.optimize(objective,n_trials=budget)
```

```python
# Measure end time
end_time = time.time()

# Calculate training time
training_time = end_time - start_time
print(study.best_params)
print(-study.best_value)
print(training_time)
```

Summing up all the execution times from the previous models, the total computational time comprised 369.946373 seconds. With Optuna, the total amount of time was 307.9956908226013 seconds. This shows the capabilities of Optuna in evaluating different models and hyperparameter combinations in an effective manner. In both cases, trying every model separately by ourselves and directly with Optuna, the best method was Gradient Boosting.

### 5.3. YellowBrick Visualizations

Once the final model was trained, we decided to look for a library that allowed us to visualize some of the training aspects for model interpretability. For this purpose, we decided to try the "Yellowbrick" library (https://www.scikit-yb.org/en/latest/), a visualization library for machine learning built on top of Scikit-Learn. It extends the Scikit-Learn API to provide a set of visual diagnostic tools called "Visualizers" that help users understand and interpret the behavior of machine learning models.


```python
import pickle
from yellowbrick.model_selection import FeatureImportances

# Load the saved pipeline
with open('final_model2.pkl', 'rb') as file:
    loaded_pipeline = pickle.load(file)

# Get the feature names from X
feature_names = X.columns
```

```python
# Visualize feature importances
viz = FeatureImportances(loaded_pipeline.named_steps['gb'], labels=feature_names, relative=False)
viz.fit(X, y)  
viz.show()  
```

```python
top_n = 10
# Access the feature importances from the fitted model
feature_importances = loaded_pipeline.named_steps['gb'].feature_importances_

# Get the indices of the top N features
top_n_indices = feature_importances.argsort()[-top_n:][::-1]

# Get the names of the top N features
top_n_feature_names = X.columns[top_n_indices]

# Print or use top N feature names as needed
print(top_n_feature_names)
```

These are not the most important variables but the ones more critical for the model, at first we intuitively assumed that the most important features would be related to the Sotavento location. However, this library allows us to see that the 10 most important features are:
- sp.24: Surface pressure at location 24
- stl1.1: Soil temperature level 1
- sp.23: Surface pressure at location 23
- sp.25: Surface pressure at location 25
- sp.22: Surface pressure at location 22
- v10n.17: Neutral wind at 10 m v-component at location 17.
- u10n.22: Neutral wind at 10 m u-component at location 22.
- v10n.18:  Neutral wind at 10 m v-component at location 18.
- cape.7: Convective available potential energy at location 7.
- lai_lv.13: Leaf area index, low vegetation at Sotavento

```python
from yellowbrick.regressor import ResidualsPlot

# Create the residuals plot using the already fitted pipeline
viz = ResidualsPlot(loaded_pipeline)
viz.score(X, y)  # Use the score method to evaluate residuals

# Show the residuals plot
viz.show()
```

When the distribution of the residuals of a prediction from a model, such as a Gradient Boosting Regressor, follows a normal distribution or closely resembles a Gaussian distribution, it generally indicates that the model's assumptions are well-met, and the model is capturing the underlying patterns in the data effectively. However, in this case, the positive mean of residuals indicates that, on average, our model tends to underestimate the target variable. This could be a sign that there are certain patterns or relationships in the data that the model is not capturing. Furthermore, the residual distribution shows that the model is heteroscedastic, which means that the variance of the residuals is not constant along the predicted values.
